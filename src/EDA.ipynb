```python
freq_df = pd.DataFrame()
freq_df['fft_mean'] = train_df.apply(lambda x: SignalOps.fft_components(x.values)[1].mean(), axis=1)
freq_df['fft_std'] = train_df.apply(lambda x: SignalOps.fft_components(x.values)[1].std(), axis=1)
freq_df['phase_mean'] = train_df.apply(lambda x: SignalOps.fft_components(x.values)[2].mean(), axis=1)
freq_df['phase_std'] = train_df.apply(lambda x: SignalOps.fft_components(x.values)[2].std(), axis=1)
```

```python
# 1 for has the disease, 0 for does not have the disease, -1 for unlabeled
train_df['target'] = -1
train_df.iloc[0:500, -1] = 1
train_df.iloc[500:1000, -1] = 0
freq_df['target'] = train_df['target']

train_df['target'].value_counts()
```

```python
labeled_train_df = train_df[train_df['target'] != -1]
labeled_freq_df = freq_df[freq_df['target'] != -1]

target = labeled_train_df['target']
labeled_train_df = labeled_train_df.drop(columns='target')
labeled_freq_df = labeled_freq_df.drop(columns='target')
labeled_train_df.shape
```

```python
WINDOW = 200
sample_healthy_signal = labeled_train_df.loc[500]
sample_ill_signal = labeled_train_df.loc[0]
healthy_sample = (sample_healthy_signal, SignalOps.moving_average(sample_healthy_signal, window_length=WINDOW))
ill_sample = (sample_ill_signal, SignalOps.moving_average(sample_ill_signal, window_length=WINDOW))
SignalPlot.binary_class_differential_plot(healthy_sample, ill_sample, title='Original Signal vs Moving Average Signal', label1='Original Signal', label2='Moving Average Signal')
```

```python
NPERSEG = 1024
labeled_freq_df['peaks'] = labeled_train_df.apply(lambda x: SignalFeatures.peaks(x, x.mean() + x.std()).shape[0], axis=1)
labeled_freq_df['ps_density_mean'] = labeled_train_df.apply(lambda x: SignalFeatures.ps_density(x, nperseg=NPERSEG)[1].mean(), axis=1)
labeled_freq_df['ps_density_std'] = labeled_train_df.apply(lambda x: SignalFeatures.ps_density(x, nperseg=NPERSEG)[1].std(), axis=1)
labeled_freq_df['entropy'] = labeled_freq_df.apply(lambda x: SignalFeatures.entropy(x), axis=1)

SignalPlot.signal_feature_plot(labeled_freq_df)
```
